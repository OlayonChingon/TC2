```python
$$ 1) V_3G_3 = (V_1-V_3) sC $$

$$ 2) (V_3-V_2) G_2 = (V_1-V_3) G_1 $$

##

$ De 2 : V_3G_2-V_2 = V_1G_1-V_3G_1 $

$ V_3(G_2+G_1) = V_1G_1 + V_2G_2 $

$$ 3) V_3=\frac{V_1G_1+V_2G_2}{G_2+G_1} $$

##
3 en 1: 

$\frac{V_1G_1+V_2G_2}{G_2+G_1} G_3 = V_1sC-(\frac{V_1G_1+V_2G_2}{G_2+G_1}) sC$

$V_1G_3G_1+V_2G_2G_3=V_1(G_2+G_1)=[V_1(G_2+G_1)-V_1G_1-V_2G_2] sC$

operando algebraicamente:

$\frac{V_2}{V_1} = -\frac{[G_3G_1-(G_2+G_1)sC+G_1sC]}{G_2G_3+G_2sC}$


$$\frac{V_2}{V_1} = \frac{s-\frac{G_3G_1}{G_2c}}{s+\frac{G_3}{C}}$$

$$\frac{V_2}{V_1} = \frac{s-\frac{R_2}{R_1}\frac{1}{R_3C}}{s+\frac{1}{R_3C}}$$


Considerando:

$R_1=R_2$

$ùúî_0=\frac{1}{R_3C}$

$$\frac{V_2}{V_1} = \frac{s-ùúî_0}{s+ùúî_0}$$

Falta analisis de modulo y fase
```

```python
#Created on Mon Apr 10 07:01:43 2024

#Tarea Semanal 1 R4001 2024

#@author: Fabian Villada




from scipy.signal import TransferFunction
import numpy as np
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot
import matplotlib.pyplot as plt

my_tf = TransferFunction([1, -1], [1, 1])

fig1, axs = bodePlot(my_tf, fig_id=1)  # Obtener la figura y las subtramas

axs[0].set_ylim(-1, 1)  # Establecer l√≠mites del eje y en la subtrama 0

pzmap(my_tf, fig_id=2)
GroupDelay(my_tf, fig_id=3)

plt.show()
```
